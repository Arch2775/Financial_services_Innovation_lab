```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot financial metrics (1.csv)
import matplotlib.pyplot as plt

def plot_financial_metrics(data):
    plt.figure(figsize=(6, 4))
    plt.plot(data['Year'], data['Net Sales'], label='Net Sales', marker='o')
    plt.plot(data['Year'], data['Operating Expenses'], label='Operating Expenses', marker='o')
    plt.plot(data['Year'], data['Cash Flow from Operations'], label='Cash Flow from Operations', marker='o')
    plt.xlabel('Year')
    plt.ylabel('Amount in Millions')
    plt.title('Financial Metrics Over Time')
    plt.legend()
    plt.grid(True)

    plt.show()


# Function to plot profit metrics (2.csv)
def plot_profit_metrics(data):
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.bar(data['Year'], data['Gross Profit (Millions)'], color='skyblue', label='Gross Profit')
    ax.set_xlabel('Year')
    ax.set_ylabel('Gross Profit in Millions')
    ax.set_title('Gross Profit Over Time')
    # Adding text for Gross Margin
    for i, v in enumerate(data['Gross Margin']):
        ax.text(data['Year'][i], data['Gross Profit (Millions)'][i] + 3, f"{v}", ha='center')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot assets and liabilities (3.csv)
def plot_assets_liabilities(data):
    plt.figure(figsize=(6, 4))
    plt.plot(data['Year'], data['Assets - Current'], label='Current Assets', marker='o')
    plt.plot(data['Year'], data['Assets - Fixed'], label='Fixed Assets', marker='o')
    plt.plot(data['Year'], data['Liabilities - Current'], label='Current Liabilities', marker='o')
    plt.plot(data['Year'], data['Liabilities - Long-Term Debt'], label='Long-Term Debt', marker='o')
    plt.xlabel('Year')
    plt.ylabel('Amount in Millions')
    plt.title('Assets and Liabilities Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
data1 = pd.read_csv('1.csv')
data2 = pd.read_csv('2.csv')
data3 = pd.read_csv('3.csv')
```

```python
data2.head()
```

```python
# Plot data
plot_financial_metrics(data1)
```

```python
plot_profit_metrics(data2)
```

```python
plot_assets_liabilities(data3)
```

```python
import matplotlib.pyplot as plt


def plot_financial_metrics_1(data):
    # Drop rows where any of the specified columns have NaN values
    clean_data = data.dropna(subset=['Revenue (Millions)', 'Operating Expenses (Millions)', 'Operating Cash Flow (Millions)'])

    plt.figure(figsize=(6, 4))
    plt.plot(clean_data['Year'], clean_data['Revenue (Millions)'], label='Revenue', marker='o')
    plt.plot(clean_data['Year'], clean_data['Operating Expenses (Millions)'], label='Operating Expenses', marker='o')
    plt.plot(clean_data['Year'], clean_data['Operating Cash Flow (Millions)'], label='Operating Cash Flow', marker='o')
    plt.xlabel('Year')
    plt.ylabel('Amount in Millions')
    plt.title('Financial Metrics Over Time')
    plt.legend()
    plt.grid(True)

    plt.show()



def plot_profit_metrics_1(data):
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.bar(data['Year'], data['Gross Margin'], color='skyblue', label='Gross Margin')
    ax.set_xlabel('Year')
    ax.set_ylabel('Gross Margin in Millions')
    ax.set_title('Gross Margin Over Time')
```

```python
plot_financial_metrics_1(data1a)
```

```python
plot_profit_metrics_1(data1b)
```
